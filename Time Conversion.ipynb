```python
#!/bin/python3

import os
import sys

s = input()

if s[len(s)-2:]=='PM':
    if int(s[:2])+12 <= 23:
        print(str(int(s[0:2])+12) + s[2:len(s)-2])
    else:
        print('12' + s[2:len(s)-2])
else:
    if int(s[0:2])==12:
        print('00' + s[2:len(s)-2])
    else:
        print(s[0:2] + s[2:len(s)-2])
```
