```python
#!/bin/python3

import math
import os
import random
import re
import sys

arr_count = int(input().strip())

arr = list(map(int, input().rstrip().split()))

max = 0
max_dict = {}

for i in set(arr):
    max_dict[i] = arr.count(i)

max = -1
ans = -1
for i,j in zip(max_dict.keys(), max_dict.values()):
    if max < j:
        if max != j:
            ans = i
        max = j

print(ans)
```
