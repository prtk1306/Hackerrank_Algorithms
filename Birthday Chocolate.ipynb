```python
#!/bin/python3

import math
import os
import random
import re
import sys

n = int(input().strip())

s = list(map(int, input().rstrip().split()))

dm = input().rstrip().split()

d = int(dm[0])

m = int(dm[1])

count = 0

for i in range(len(s)-m+1):
    if sum(s[i:i+m]) == d:
        count += 1
print(count)
```
