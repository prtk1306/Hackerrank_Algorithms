```python
#!/bin/python3

import math
import os
import random
import re
import sys

nk = input().split()

n = int(nk[0])

k = int(nk[1])

ar = list(map(int, input().rstrip().split()))

count = 0

for i in range(len(ar)):
    for j in range(i+1, len(ar)):
        if (((ar[i] + ar[j]) % k) == 0) and i != j:
            count += 1
print(count)
```
