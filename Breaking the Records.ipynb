```python
#!/bin/python3

import math
import os
import random
import re
import sys


n = int(input())

scores = list(map(int, input().rstrip().split()))

max_count = 0
min_count = 0

max_score = scores[0]
min_score = scores[0]

for i in scores[1:]:
    if max_score < i:
        max_score = i
        max_count += 1
        
    if min_score > i:
        min_score = i
        min_count += 1
    
print(str(max_count) + ' ' + str(min_count))
```
